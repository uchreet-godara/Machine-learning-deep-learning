```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib._image as mpimg
import seaborn as sns
import pandas as pd
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
import plotly.offline as py
py.init_notebook_mode(connected = True)
```

```python
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
trainX = pd.read_csv('../input/train.csv')
```

```python
testX = pd.read_csv('../input/test.csv')
```

```python
trainX.shape
```

```python
X = np.array(trainX)
```

```python
X = X[:6000]
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Xstd = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components=5)
```

```python
pca.fit(Xstd)
```

```python
X5 = pca.transform(Xstd)
```

```python
Target = np.array(trainX.iloc[:,0])
```

```python
tt = go.Scatter(x=X5[:,0],y=X5[:,1],
               mode = 'markers',
               text = Target,
               showlegend = False,
               marker =dict(
               size = 8,
               color = Target,
               colorscale = 'Jet',
               showscale = True,
               line = dict(
               width = 2,
               color = 'rgb(255,255,255)'),
               opacity = .8
               ))

lyot = go.Layout(
                title = 'PCA',
                hovermode = 'closest',
                 xaxis = dict(
                 title = 'fct',
                 ticklen = 5,
                 gridwidth = 2,),
  
                 yaxis = dict(title = 'Sct',
                      ticklen = 5,
                      gridwidth = 2,
                             ),
                 showlegend = False
                )

data = [tt]
```

```python
fig = dict(data=data, layout=lyot)
py.iplot(fig, filename='styled-scatter')
```
